Q.No-03    What is App routing in Flask? Why do we use app routes?

Ans :-

`App routing` refers to the process of mapping URLs (Uniform Resource Locators) to functions in your application. It allows you to define different URL patterns and associate them with specific functions or views in your Flask application. These URL patterns are called routes.

Routes are defined using the `@app.route()` decorator in Flask. The decorator specifies the URL pattern or endpoint that will trigger the associated function when accessed by a client.

    Here's an example to illustrate the concept:

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return 'Hello, World!'

@app.route('/about')
def about():
    return 'About page'

if __name__ == '__main__':
    app.run()
```

By using app routes, you can create different endpoints for various parts of your application. For example, you can have separate routes for handling user authentication, displaying different pages or views, handling form submissions, and so on. Each route can be associated with a specific function that performs the necessary logic and returns an appropriate response.

App routing is crucial in Flask (and other web frameworks) as it helps organize your application's functionality and provides a clean way to handle different URLs. It allows you to define the behavior of your application based on the requested URLs, making it easier to build and maintain web applications.

Q.No-04    Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:

    Company Name: ABC Corporation

    Location: India

    Contact Detail: 999-999-9999 

Ans :-

````python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

@app.route('/')
def company_details():
    return '''<h1>Company Name: ABC Corporation</h1>
              Location: India
              Contact Detail: 999-999-9999'''

if __name__ == '__main__':
    app.run()
````

Q.No-05    What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the `url_for()` function is used for URL building. It generates a URL for the specified endpoint with the given arguments. This function takes the endpoint name as its first argument and additional keyword arguments for the endpoint's variables.

Here's a Python code example to demonstrate the working of the `url_for()` function in Flask:

````python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    # Using url_for() to generate a URL for the 'hello' endpoint
    hello_url = url_for('hello')
    return f'Hello!<a href="{hello_url}"><h1>Click here</h1></a> to go to the Hello page.'

@app.route('/hello')
def hello():
    return "<h1>Hello, World!</h1>"

if __name__ == '__main__':
    app.run()
````
